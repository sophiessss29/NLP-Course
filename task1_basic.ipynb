```python
import nltk
```

```python
import spacy
```

```python
!wget -O data.txt https://raw.githubusercontent.com/vifirsanova/compling/refs/heads/main/tasks/task1/data.txt
```

```python
with open("data.txt", "r", encoding="utf-8") as file:
    data = file.read()
```

```python
import string
```

```python
nltk.download('punkt')
```

```python
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(data)
```

```python
cleaned_tokens = [word for word in tokens if word not in string.punctuation]
```

```python
cleaned_text = ' '.join(cleaned_tokens)
```

```python
print("Текст без пунктуации:", cleaned_text)
```

```python
cleaned_text = cleaned_text.lower()
```

```python
print("Текст в нижнем регистре:", cleaned_text)
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
filtered_words = [word for word in tokens if word.lower() not in stop_words and word not in string.punctuation]
```

```python
print("Текст без стоп-слов:", filtered_words)
```

```python
filtered_text = ' '.join(filtered_words)
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(filtered_text)
```

```python
lemmatized_words = ' '.join([token.lemma_ for token in doc if not token.is_punct])
```

```python
print("Лемматизированный текст:", lemmatized_words)
```

```python
print("Итоговый текст:", lemmatized_words)
```
