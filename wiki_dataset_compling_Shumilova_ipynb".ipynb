```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/Stanford_University'
r_wiki = requests.get(url=url)
```

```python
r_wiki.text
```

```python
soup_wiki = BeautifulSoup(r_wiki.text, 'html.parser')
soup_wiki
```

```python
pars_wiki = soup_wiki.find_all('p')
pars_wiki
```

```python
texts_wiki = [text.get_text() for text in pars_wiki]
texts_wiki
```

```python
texts_wiki[1]
```

```python
texts_wiki = '\n'.join([x for x in texts_wiki])
```

```python
data = pars_wiki
with open('data.txt', 'w') as file:
    file.write(str(data))
```

```python
data = texts_wiki
with open('data.txt', 'w') as file:
    file.write(str(data))
```
